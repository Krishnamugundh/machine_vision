```python
from imageai.Classification import ImageClassification
import os
import cv2

execution_path = os.getcwd()
```

```python
img1 = cv2.imread("/content/photo-1589828155685-83225f7d91f3.jpg")
```

```python
from imageai.Detection import ObjectDetection
import os

execution_path = os.getcwd()
```

```python
detector = ObjectDetection()
detector.setModelTypeAsTinyYOLOv3()
detector.setModelPath( os.path.join(execution_path , "/content/drive/MyDrive/Colab Notebooks/Machine Vision/tiny-yolov3.pt"))
```

```python
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "/content/photo-1589828155685-83225f7d91f3.jpg"), output_image_path=os.path.join(execution_path , "imagenew.jpg"), minimum_percentage_probability=30)
```

```python
for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"] )
    print("--------------------------------")
```

```python
for i in detections:
  x1,y1,x2,y2 = i["box_points"]
  cv2.rectangle(img1, (x1, y1), (x2, y2), (0, 255, 0), 2)
  if(i['percentage_probability']<90):
    name = "Defected"
  else:
    name = "Perfect"
  cv2.putText(
    img1,
    name,
    (int(x1), int(y1) - 10),
    fontFace = cv2.FONT_HERSHEY_SIMPLEX,
    fontScale = 0.6,
    color = (200,200,200),
    thickness=2
)
```
